```python
# Removing Outliers
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# Checking the outliers using formula True means it is an outlier false means valid not an outlier. 

print(df<(Q1-1.5*IQR))|(df>(Q3+1.5*IQR))
```

```python
# all outliers are removed
# removing outliers
df = df[~((df<(Q1-1.5*IQR))|(df>(Q3+1.5*IQR))).any(axis=1)]
```

```python
df.shape
```

```python
# Feature Engineering
df['No_of_year'] = 2020-df['Year'] # creating new year column
```

```python
# handle the categorical columns or variable.
col1 = ['Fuel_Type','Transmission','Seller_Type']
df = pd.get_dummies(df,columns=col1,drop_first=True)
```

```python
# droping unwanted columns
drop_var =['Car_Name','Year']
df.drop(drop_var,axis=1,inplace=True)
```

```python
# Checking the corelation between the variables
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True,cmap='RdYlGn')
plt.show()
```

```python
df.head(3)
```

```python
x = df.iloc[:,1:] # splitting the data
y = df.iloc[:,0]
```

```python
from sklearn.ensemble import ExtraTreesRegressor
model= ExtraTreesRegressor()
model.fit(x,y)
```

```python
print(model.feature_importances_)
```

```python
feat_importances = pd.Series(model.feature_importances_, index=x.columns).sort_values(ascending=False)
feat_importances.plot(kind='barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
y_pred_train = regressor.predict(x_train)                    # For train data
print('r2_score {}'.format(r2_score(y_train,y_pred_train)))
print('mean_squared_error {}'.format(mean_squared_error(y_train,y_pred_train)))
print('RMSE {}'.format(np.sqrt(mean_squared_error(y_train,y_pred_train))))
print('mean_absolute_error {}'.format(mean_absolute_error(y_train,y_pred_train)))
```

```python
y_pred_test = regressor.predict(x_test)                    # For test data
print('r2_score {}'.format(r2_score(y_test,y_pred_test)))
print('mean_squared_error {}'.format(mean_squared_error(y_test,y_pred_test)))
print('RMSE {}'.format(np.sqrt(mean_squared_error(y_test,y_pred_test))))
print('mean_absolute_error {}'.format(mean_absolute_error(y_test,y_pred_test)))
```
